```python
import numpy as np
import pandas as pd
```

```python
all_train_X = np.load("../train_data/X_train_kaggle.npy")
```

```python
all_train_X.shape
```

```python
# if the values in the data are in the same order as mentioned in competition info...
train_X_by_attribute = {
    "Orientation_X" : all_train_X[:,0,:],
    "Orientation_Y" : all_train_X[:,1,:],
    "Orientation_Z" : all_train_X[:,2,:],
    "Orientation_W" : all_train_X[:,3,:],
    "AngularVelocity_X" : all_train_X[:,4,:],
    "AngularVelocity_Y" : all_train_X[:,5,:],
    "AngularVelocity_Z" : all_train_X[:,6,:],
    "LinearAcceleration_X" : all_train_X[:,7,:],
    "LinearAcceleration_Y" : all_train_X[:,8,:],
    "LinearAcceleration_Z" : all_train_X[:,9,:],
}
linZ = train_X_by_attribute.get("LinearAcceleration_Z")
linZ.shape
```

```python
# (1) X_mean_var will have 10 lists, each holding values of one attribute -- dimension 1
# (2) all lists of (1) will hold 1703 cases -- dimension 2
# (3) all lists of (2) will hold a list of two values [mean, variance] -- "dimension 3"

# [10]         [1703]   [2]
# attributes   cases    values -> [mean, variance]
att_num = 10
case_num = 1703
X_mean_var = []

for i in range(att_num):
    this_attribute = []
    cur = all_train_X[:,i,:]
    
    for j in range(case_num):
        this_mean = np.mean(cur[j], axis=0)
        this_variance = np.var(cur[j], axis=0)
        this_case = [this_mean, this_variance]
        this_attribute.append(this_case)
        
    X_mean_var.append(this_attribute)
    
X_train_mean_var = X_mean_var
```

```python
# mean and standard deviation here
att_num = 10
case_num = 1703
X_train_mean_std = []

for i in range(att_num):
    this_attribute = []
    cur = all_train_X[:,i,:]
    
    for j in range(case_num):
        this_mean = np.mean(cur[j], axis=0)
        this_std = np.std(cur[j], axis=0)
        this_case = [this_mean, this_std]
        this_attribute.append(this_case)
        
    X_train_mean_std.append(this_attribute)
```

```python
print(len(X_train_mean_std))
print(len(X_train_mean_std[-1]))
print(len(X_train_mean_std[-1][-1]))
```

```python
# create a pandas df for all the attributes separately. One df from one attribute and then later combine them as desired
# for testing different attribute combinations
dfs_mean_std = []

# "Orientation_X"
df_Xtrain_mean_std_orientationX = pd.DataFrame(X_train_mean_std[0], columns = ['OrientationX_mean', 'OrientationX_std',])
dfs_mean_std.append(df_Xtrain_mean_std_orientationX)

# "Orientation_Y"
df_Xtrain_mean_std_orientationY = pd.DataFrame(X_train_mean_std[1], columns = ['OrientationY_mean', 'OrientationY_std',])
dfs_mean_std.append(df_Xtrain_mean_std_orientationY)

# "Orientation_Z"
df_Xtrain_mean_std_orientationZ = pd.DataFrame(X_train_mean_std[2], columns = ['OrientationZ_mean', 'OrientationZ_std',])
dfs_mean_std.append(df_Xtrain_mean_std_orientationZ)

# "Orientation_W"
df_Xtrain_mean_std_orientationW = pd.DataFrame(X_train_mean_std[3], columns = ['OrientationW_mean', 'OrientationW_std',])
dfs_mean_std.append(df_Xtrain_mean_std_orientationW)

# "AngularVelocity_X"
df_Xtrain_mean_std_angularVelocityX = pd.DataFrame(X_train_mean_std[4], columns = ['AngularVelocityX_mean', 'AngularVelocityX_std',])
dfs_mean_std.append(df_Xtrain_mean_std_angularVelocityX)

# "AngularVelocity_Y"
df_Xtrain_mean_std_angularVelocityY = pd.DataFrame(X_train_mean_std[5], columns = ['AngularVelocityY_mean', 'AngularVelocityY_std',])
dfs_mean_std.append(df_Xtrain_mean_std_angularVelocityY)

# "AngularVelocity_Z"
df_Xtrain_mean_std_angularVelocityZ = pd.DataFrame(X_train_mean_std[6], columns = ['AngularVelocityZ_mean', 'AngularVelocityZ_std',])
dfs_mean_std.append(df_Xtrain_mean_std_angularVelocityZ)

# "LinearAcceleration_X"
df_Xtrain_mean_std_linearAccelerationX = pd.DataFrame(X_train_mean_std[7], columns = ['LinearAccelerationX_mean', 'LinearAccelerationX_std',])
dfs_mean_std.append(df_Xtrain_mean_std_linearAccelerationX)

# "LinearAcceleration_Y"
df_Xtrain_mean_std_linearAccelerationY = pd.DataFrame(X_train_mean_std[8], columns = ['LinearAccelerationY_mean', 'LinearAccelerationY_std',])
dfs_mean_std.append(df_Xtrain_mean_std_linearAccelerationY)

# "LinearAcceleration_Z"
df_Xtrain_mean_std_linearAccelerationZ = pd.DataFrame(X_train_mean_std[9], columns = ['LinearAccelerationZ_mean', 'LinearAccelerationZ_std',])
dfs_mean_std.append(df_Xtrain_mean_std_linearAccelerationZ)
```
